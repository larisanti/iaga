## Lista de Exercícios

```python
# 1. Cálculo de desconto de 10%
valor = float(input())

desconto = valor * 0.1
valor_final = valor - desconto

print(f"Valor do Produto  = R$ {valor:.2f}")
print(f"Valor do Desconto = R$ {desconto:.2f}")
print(f"Valor Final       = R$ {valor_final:.2f}")
```

```python
# 2. Busca sequencial em lista de inteiros
a = [10, 2, 7, 8, 5, 3, 22, 17, 18]

item = int(input())
#print(a[1])
print({i: a[i] for i in range(len(a))})

for i in range(len(a)):

    if item == a[i]:
        print(f"Item: {item}, foi encontrado na posição {i}")
        break
else:
    print(f"Item: {item}, não foi encontrado.")
```

```python
# 3. Cálculo de desconto condicional (10% ou 15%)
valor = float(input())
print(f"Valor do Produto = R$ {valor:.2f}")

desconto10 = valor * 0.10
desconto15 = valor * 0.15

#print(f"{valor:.2f}, {desconto10:.2f}, {desconto15:.2f}")

if valor < 5000:
    print(f"Valor do Desconto (10%) = R$ {desconto10:.2f}")
    print(f"Valor Final = R$ {valor - desconto10:.2f}")
else:
    valor > 5000
    print(f"Valor do Desconto (15%) = R$ {desconto15:.2f}")
    print(f"Valor Final = R$ {valor - desconto15:.2f}")
```

```python
# 4. Geração de matriz com caracteres * e @
n = int(input())

for linha in range(n):
    for coluna in range(n):
        if (linha == 0) or (linha == n - 1) or (coluna == 0) or (coluna == n - 1):
            print("*", end=" ")
        else:
            print("@", end=" ")
    print() 
```

```python
# 5. Repetição de string
string = input("")

for i in range(3):
    print(" ".join(string))
```

```python
# 6. Contagem de palavras em uma frase
string = input("")

if string.endswith('.'):
    string = string[:-1]
    
palavras = string.split()

print(f"Existem {len(palavras)} palavras, são elas:")
print()
for i, palavra in enumerate(palavras, start=1):
    print(f"{i}a. palavra = {palavra}")
```

```python
# 7. Alternância de valores par/ímpar em lista
n = int(input())
l = []

for i in range(n):
    l.append(i % 2)

#print(i)

resultados = ", ".join(str(x) for x in l)
print("{" + resultados + "}")
```

```python
# 8. Ordenação de lista e identificação de valores
elementos = []

for i in range(7):
    item = int(input())
    elementos.append(item)

elementos.sort()

for i in range(7):
    print(f"x[{i}] = {elementos[i]}")
print()
print(f"Menor elemento, x[0] = {elementos[0]}")
print(f"Maior elemento, x[6] = {elementos[6]}")
```

```python
# 9. Preenchimento de lista com progressão linear
a = int(input())

for i in range(10):
    resultado = a + i
    print(f"x[{i}] = {a} + {i} = {resultado}")
```

```python
# 10. Classificação de elementos em lista (negativo/nulo/positivo)
n = int(input())

lista1 = []
for i in range(n):
    lista1.append(int(input()))

lista2 = []
for x in lista1:
    if x < 0:
        lista2.append(-1)
    elif x == 0:
        lista2.append(0)
    else:
        lista2.append(1)

print(lista1)
print(lista2)
```

```python
# 11. Antecessor e sucessor
n = int(input())

print(f"antecessor: {n - 1}")
print(f"numero....: {n}")
print(f"sucessor..: {n + 1}")
```

```python
# 12. Print do valor do meio entre três números
a = int(input())
b = int(input())
c = int(input())

valores = [a, b, c]
valores.sort()

print(f"meio = {valores[1]}")
```

```python
# 13. Matriz figura com diagonal (caractere @)
n = int(input())

for i in range(n):
    for j in range(n):
        if i == j:
            print("@", end=" ")
        else:
            print(".", end=" ")
    print()
```

```python
# 14. Vetor com inteiros (intervalo 0 a 99)
a = []

for n in range(8):
    a.append(int(input()))

for i in range(8):
    print(f"a[{i}] = {a[i]}")

menor = min(a)
print()
print(f"Menor valor = {menor}")
```

```python
# 15. Inversão de caracteres de uma string
string = input()

print(f"Entrada: {string}")

invertida = string[::-1]
print(f"Saída..: {invertida}")
```
