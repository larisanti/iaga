### Notes
- Clustering é uma técnica de aprendizado não supervisionado que agrupa dados por similaridade, sem rótulos predefinidos. 
- O objetivo é maximizar a homogeneidade interna e a heterogeneidade entre grupos.
- Normalização: atributos com escalas diferentes distorcem a distância euclidiana. 
- Métodos pra normalizar: 
    - Min-Max (sensível a outliers)
    - Log (bom para dados financeiros)
    - Valor Máximo (risco de achatamento)
-  Algoritmo K-Means para de iterar quando não há mudança nos centróides entre uma iteração e a seguinte (os clusters se estabilizaram)
- K-means falha em clusters de formato alongado.
- Distância Euclidiana: mede a distância em linha reta entre dois pontos e é a métrica padrão do K-Means para definir fronteiras de cluster
- SSE + Cotovelo = qualidade medida pela soma dos quadrados dos erros. 
- SSE = 0 indica overfitting

### Etapas - Distância Euclidiana
1. Subtrai as coordenadas ponto a ponto em cada dimensão
2. Eleva cada diferença ao quadrado
3. Soma todos os quadrados
4. Extrai a raiz quadrada do total

Fórmulas:
2D ->  d = √[(p₁−q₁)² + (p₂−q₂)²]
ND -> d = √[Σ(xᵢ−yᵢ)²]

### Etapas - K-means
1. Escolher K centros iniciais aleatoriamente
2. Atribuir cada objeto ao centro mais próximo
3. Recalcular os centróides
3. Repetir até não haver mudança nos centros
