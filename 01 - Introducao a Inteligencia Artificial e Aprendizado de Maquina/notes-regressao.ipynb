### Notes
- Escolher este modelo se a resposta que busco é um número de valor contínuo
- A regressão linear prevê valores numéricos contínuos (y) a partir de variáveis preditoras/dados históricos (x)
- Equação da reta: y = a * x + b
                        ^ a = taxa de mudança | b = valor da base
- R² (0-1) -> mede quanto x explica quanto y muda
- R² bom depende da área, para engenharia é geralmente acima de 0,90
- Outliers distorcem a reta
- Resíduos em formato U indicam relação não linear -> usar regressão polinomial ou deep learning
- Correlação ≠ causalidade
- O modelo é uma aproximação estatística, o valor está na interpretação, não no cálculo
- Relação:
    1. Positiva: quando X aumenta, Y também aumenta (mesma direção)
    2. Negativa: quando X aumenta, Y diminui (direções opostas)
- Ausência de relação: quando os pontos estão espalhados pelo gráfico como uma nuvem de caos

### Etapas
