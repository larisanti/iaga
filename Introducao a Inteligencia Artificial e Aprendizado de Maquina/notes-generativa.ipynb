### Notes
- IA Tradicional (padrões em dados estruturados, ex: K-Means)
- IA Generativa (criação de conteúdo novo via linguagem)
- Aprende padrões.
- Multimodal: produz texto, imagem, etc
- Gera materiais novos a partir de padrões aprendidos.
- Tokens -> pedaços de texto (~750 palavras/1000 tokens)
- Embeddings -> palavras viram vetores; sentidos próximos = coordenadas próximas.
- Atenção -> foco contextual (distinguir "banco" financeiro vs assento). 
- Palpite probabilístico -> a IA "chuta" a próxima palavra mais provável, sem conhecimento consciente.
- RAG: força a IA a ancorar respostas em documentos fornecidos para reduzir alucinações.
- Bom prompt: papel, contexto, tarefa, regra, formato.
- "A incerteza é uma virtude científica. Não deixe a IA apagar suas dúvidas."
