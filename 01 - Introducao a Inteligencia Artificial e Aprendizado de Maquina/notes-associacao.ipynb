### Notes
- Associação é uma técnica de aprendizado não supervisionado (sem variável alvo).
- É diferente da clusterização, que agrupa perfis semelhantes.
- Mapeia conexões entre objetos: o que vai com o quê (market basket analysis).
- Não implica causalidade, cabe ao analista conferir significado.
- Métricas:
    - suporte: quão comum é o padrão na base total?
    - confiança: se x ocorreu, qual a chance de y? (denominador = ocorrências de x)
    - lift: a regra traz conhecimento novo?
- Métrica de confiança: se X então Y (antecedente -> consequente). 
- Suporte descarta regras sem escala.
- Lift remove viés de itens já populares.
- Apriori: algoritmo que resolve a explosão combinatória
- Poda anti-monotônica: se um itemset não atinge o suporte mínimo, todos os seus superconjuntos são descartados automaticamente.

### Etapas (Excel)
1. 1x1 -> suporte individual de cada item: eliminar < 8
2. 2x2 -> cruzar e contar coocorências: mantém pares com suporte > 0.30 (COUNTIFS)
3. 3x3 -> montar trios: se suporte > 0.30, então calcular confiança e lift
