### Notes
- A classificação é a técnica que atribui uma categoria a um objeto com base em suas características, como aprovar ou reprovar um crédito, ou diagnosticar um paciente como doente ou sadio
- Os dados de entrada do modelo são os atributos históricos (como idade e renda) e a saída será sempre uma classe (como Sim ou Não)
- Contextos: diagnósticos na medicina, detecção de fraudes para a área de segurança e filtros de spam
- A máquina pode tomar decisões por meio de regras, probabilidades, delimitação de fronteiras ou por métricas de proximidade
- Árvore de decisão: modelo toma decisões por meio de perguntas lógicas sequenciais e possui altíssima interpretabilidade
    ^ pergunta (nó) -> resposta (ramo) -> decisão (folha)
- Regressão logística: curva Sigmóide pra converter valores em probabilidade (0-1)
- SVM: algoritmo busca criar a fronteira de separação mais larga possível
- KNN: classe majoritária entre vizinhos mais próximos
- Redes neurais: camadas que extraem padrões complexos (base da IA Generativa)
- Dependendo do contexto taxa de acerto de 95% pode ser classificada como um desastre
- Matriz de confusão: avaliação do modelo com quadrantes de sucesso
    - Verdadeiro Positivo (VP) / Verdadeiro Negativo (VN): acerto
    - Falso Positivo (FP - Erro Tipo I): alarme falso
    - Falso Negativo (FN - Erro Tipo II): omissão perigosa
- Acurácia: enganosa em dados desbalanceados.
- Precisão: priorizar quando FP é caro (ex: bloqueio indevido de cartão).
- Recall: priorizar quando FN é perigoso (ex: doente sem diagnóstico).
- F1-Score: média harmônica entre Precisão e Recall, melhor para dados desbalanceados.
