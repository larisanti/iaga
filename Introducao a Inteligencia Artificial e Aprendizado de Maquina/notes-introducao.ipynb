### Notes
- Achei interessante o exemplo da professora sobre o Deep Blue e que nem toda IA aprende.
- A escolha certa da técnica é mais importante do que dominar a implementação.
- A maior parte das aplicações nas organizações são de IA tradicional.
- Inteligência Artificial: capacidade de sistemas realizarem tarefas que normalmente exigiriam inteligência humana.
- Machine Learning: ramo da IA em que os sistemas aprendem padrões diretamente a partir dos dados, sem precisarem ser explicitamente programados.
- Os problemas solucionados por IA são organizados em quatro categorias conceituais de aplicação:
    1. Regressão: arquitetada para a previsão de valores numéricos contínuos (ex.: estipular o preço de um imóvel ou medir variações de temperatura);
    2. Classificação: focada na tomada de decisão baseada em categorias (ex.: definir se uma operação financeira é fraude ou não-fraude);
    3. Clusterização: utilizada para o agrupamento de dados semelhantes sem um rótulo prévio (ex.: criação de segmentos e perfis de clientes);
    4. Associação: método de relacionar padrões e frequências de coocorrência (ex.: descobrir a probabilidade de um cliente comprar o produto Y dado que ele comprou o produto X).
- IA Generativa: termo dentro do umbrella Deep Learning, focado em criar.
